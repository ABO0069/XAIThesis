```python

# ml_analysis_pipeline#

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import pickle
import joblib
import sys

# Core ML & preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
from imblearn.over_sampling import SMOTE

# Deep learning
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Explainability
import lime
import lime.lime_tabular
import shap
import dice_ml
from dice_ml import Dice

# Fairness & bias
from aif360.datasets import BinaryLabelDataset

# Causality
from dowhy import CausalModel

from llm_layer import ClinicalLLM

# Colab-specific utility
try:
    from google.colab import files
    COLAB_ENV = True
except ImportError:
    COLAB_ENV = False

warnings.filterwarnings('ignore')
np.random.seed(42)

# CONSTANTS
RANDOM_STATE = 42
TEST_SIZE = 0.2

# Define all feature columns
FEATURE_COLUMNS = [
    'age', 'gender_encoded', 'pulse_rate', 'systolic_bp', 'diastolic_bp',
    'glucose', 'height', 'weight', 'bmi', 'family_diabetes', 'hypertensive',
    'family_hypertension', 'cardiovascular_disease', 'stroke'
]

# Feature names for interpretability
FEATURE_NAMES_HUMAN = [
    'Age', 'Gender', 'Pulse Rate', 'Systolic BP', 'Diastolic BP',
    'Glucose', 'Height', 'Weight', 'BMI', 'Family Diabetes', 'Hypertensive',
    'Family Hypertension', 'Cardiovascular Disease', 'Stroke'
]

# Categorical features for DiCE/Fairness
CATEGORICAL_FEATURES = [
    'gender_encoded','family_diabetes','hypertensive',
    'family_hypertension','cardiovascular_disease','stroke'
]
# ---------------------------------

def load_and_preprocess_data(data_path=None):
    """1. Loads, encodes, resamples, splits, and scales the data."""
    print("\n1. LOADING AND PREPROCESSING DATA")
    print("-" * 40)
    df = None

    if COLAB_ENV and data_path is None:
        # Colab Mode:session file upload
        print("Colab Environment Detected. Opening file upload dialog...")
        uploaded = files.upload()
        if not uploaded:
            raise FileNotFoundError("No file uploaded in Colab session.")

        fn = next(iter(uploaded.keys()))
        df = pd.read_csv(fn)
        print(f"Uploaded file: {fn}")

    elif data_path:
        # Local Mode: Data path provided
        print(f"Reading data from provided path: {data_path}")
        df = pd.read_csv(data_path)

    else:
        # Not Colab and no data_path provided
        raise ValueError("Cannot load data. Please provide a `data_path` parameter or run in a Colab environment to prompt upload.")

    if df is not None:
        print("\nFirst 5 rows of the loaded dataset:")
        print(df.head())

    # Encode categorical and target variables
    le = LabelEncoder()

    df['gender_encoded'] = le.fit_transform(df['gender'])
    df['target'] = le.fit_transform(df['diabetic'])
    X = df[FEATURE_COLUMNS]
    y = df['target']

    print(f"\nOriginal class distribution: {np.bincount(y)}")

    # Handle class imbalance
    smote = SMOTE(random_state=RANDOM_STATE)
    X_resampled, y_resampled = smote.fit_resample(X, y)
    print(f"Resampled class distribution: {np.bincount(y_resampled)}")

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_resampled, y_resampled,
        test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y_resampled
    )

    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    return df, X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled, scaler, le

def train_and_evaluate_models(X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled):
    """2. Trains Random Forest, Logistic Regression, and Neural Network models."""
    print("\n2. TRAINING MULTIPLE MODELS")
    print("-" * 40)

    # Random Forest Model
    rf_model = RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE, max_depth=10)
    rf_model.fit(X_train, y_train)
    rf_predictions = rf_model.predict(X_test)
    rf_probabilities = rf_model.predict_proba(X_test)[:, 1]

    # Logistic Regression Model
    lr_model = LogisticRegression(random_state=RANDOM_STATE, max_iter=1000)
    lr_model.fit(X_train_scaled, y_train)
    lr_predictions = lr_model.predict(X_test_scaled)
    lr_probabilities = lr_model.predict_proba(X_test_scaled)[:, 1]

    # Neural Network Model
    nn_model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(16, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    nn_model.fit(
        X_train_scaled, y_train,
        validation_split=0.2, epochs=100, batch_size=32,
        callbacks=[early_stopping], verbose=0
    )
    nn_probabilities = nn_model.predict(X_test_scaled).flatten()
    nn_predictions = (nn_probabilities > 0.5).astype(int)

    models = {
        'Random Forest': (rf_model, rf_predictions, rf_probabilities),
        'Logistic Regression': (lr_model, lr_predictions, lr_probabilities),
        'Neural Network': (nn_model, nn_predictions, nn_probabilities)
    }

    print("Model Performance:")
    print("-" * 65)
    print(f"{'Model':20} {'Acc':>6} {'Precision':>10} {'Recall':>8} {'F1':>6} {'AUC':>6}")
    print("-" * 65)
    for name, (_, pred, prob) in models.items():
        acc = accuracy_score(y_test, pred)
        prec = precision_score(y_test, pred, zero_division=0)
        rec = recall_score(y_test, pred, zero_division=0)
        f1 = f1_score(y_test, pred)
        auc = roc_auc_score(y_test, prob)
        print(f"{name:20} - {acc:.4f} {prec:.4f} {rec:.4f} {f1:.4f} {auc:.4f}")
    print("-" * 65)
    # ----------------------------------------------------------------

    # Plot ROC Curves
    plt.figure(figsize=(8, 6))
    for name, (_, _, prob) in models.items():
        fpr, tpr, _ = roc_curve(y_test, prob)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc_score(y_test, prob):.4f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves for All Models')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return models

def setup_explainability_tools(X_train, rf_model, lr_model, X_train_scaled):
    """3. Initializes LIME and SHAP explainers."""
    print("\n3. SETTING UP EXPLAINABILITY TOOLS")
    print("-" * 40)

    # LIME Setup
    lime_explainer = lime.lime_tabular.LimeTabularExplainer(
        X_train.values,
        feature_names=FEATURE_NAMES_HUMAN,
        class_names=['No Diabetes', 'Diabetes'],
        mode='classification'
    )

    # SHAP Setup
    X_background_sample = X_train.sample(n=500, random_state=RANDOM_STATE).values.astype(float)
    shap_explainer_rf = shap.TreeExplainer(rf_model, X_background_sample)
    shap_explainer_lr = shap.LinearExplainer(lr_model, X_train_scaled)

    print("LIME and SHAP explainers initialized successfully!")
    return lime_explainer, shap_explainer_rf, shap_explainer_lr

def evaluate_fairness_metrics(X_test, y_test, models):
    """4. Evaluates fairness metrics for all models."""
    print("\n4. FAIRNESS EVALUATION")
    print("-" * 40)
    def calculate_fairness(X_test_df, y_test_series, predictions, sensitive_features=['gender_encoded', 'age']):
        results = {}
        for feature in sensitive_features:
            feature_values = X_test_df[feature].values.copy()
            if feature == 'age':
                bins = [0, 35, 60, 120]
                labels = ['Young', 'Middle', 'Old']
                groups = pd.cut(feature_values, bins=bins, labels=labels, include_lowest=True)
            else:
                groups = feature_values

            feature_result = {}
            for group in np.unique(groups):
                mask = groups == group
                if np.sum(mask) > 0:
                    group_accuracy = accuracy_score(y_test_series[mask], predictions[mask])
                    group_precision = precision_score(y_test_series[mask], predictions[mask], zero_division=0)
                    group_recall = recall_score(y_test_series[mask], predictions[mask], zero_division=0)

                    feature_result[group] = {
                        'accuracy': group_accuracy,
                        'precision': group_precision,
                        'recall': group_recall,
                        'count': np.sum(mask)
                    }

            if len(feature_result) > 1:
                first_group = list(feature_result.keys())[0]
                last_group = list(feature_result.keys())[-1]

                disparate_impact = feature_result[first_group]['accuracy'] / feature_result[last_group]['accuracy'] if feature_result[last_group]['accuracy'] > 0 else 0.0
                feature_result['disparate_impact'] = disparate_impact
                feature_result['fairness_satisfied'] = 0.8 <= disparate_impact <= 1.25

            results[feature] = feature_result
        return results

    X_test_df_reset = X_test.reset_index(drop=True)
    y_test_series_reset = y_test.reset_index(drop=True)

    print("Fairness Evaluation Results (Gender + Age):")
    fairness_results_all_models = {}
    for name, (_, pred, _) in models.items():
        fairness_results = calculate_fairness(X_test_df_reset, y_test_series_reset, pred)
        fairness_results_all_models[name] = fairness_results
        print(f"\n{name}:")
        for feature, values in fairness_results.items():
            print(f"  Feature: {feature}")
            for key, value in values.items():
                if isinstance(value, dict):
                    print(f"    {key}: Acc={value['accuracy']:.4f}, Prec={value['precision']:.4f}, Recall={value['recall']:.4f}, Count={value['count']}")
                else:
                    print(f"    {key}: {value}")
    return fairness_results_all_models

def perform_causal_analysis(df):
    """5. Performs a sample causal analysis using DoWhy."""
    print("\n5. CAUSALITY ANALYSIS")
    print("-" * 40)

    def analyze(df, treatment_var, outcome_var, confounders):
        try:
            df_for_causal = df.copy()
            if treatment_var not in df_for_causal.columns or treatment_var != 'high_glucose':
                 df_for_causal['high_glucose'] = (df_for_causal['glucose'] > df_for_causal['glucose'].median()).astype(int)

            causal_model = CausalModel(
                data=df_for_causal,
                treatment=treatment_var,
                outcome=outcome_var,
                confounders=confounders
            )
            identified_estimand = causal_model.identify_effect()
            causal_estimate = causal_model.estimate_effect(
                identified_estimand,
                method_name="backdoor.linear_regression"
            )
            print(f"Causal effect of {treatment_var} on {outcome_var}: {causal_estimate.value:.4f}")
            conf_intervals = causal_estimate.get_confidence_intervals()
            lower_bound = float(conf_intervals[0, 0])
            upper_bound = float(conf_intervals[0, 1])
            print(f"Confidence intervals: [{lower_bound:.4f}, {upper_bound:.4f}]")
            return causal_estimate

        except Exception as e:
            print(f"Causal analysis failed: {str(e)}")
            return None

    confounders = ['age', 'bmi', 'systolic_bp', 'family_diabetes']
    causal_result = analyze(df, treatment_var='high_glucose', outcome_var='target', confounders=confounders)
    return causal_result

def setup_dice_counterfactuals(df, rf_model, X_test):
    """6. Sets up and runs DiCE counterfactual explanations for an instance."""
    print("\n6. COUNTERFACTUAL EXPLANATIONS SETUP (DiCE)")
    print("-" * 40)

    # 1. Prepare df_dice
    df_dice = df.drop(columns=['gender','diabetic'])
    df_dice = df_dice[FEATURE_COLUMNS + ['target']] # Use consistent order

    # 2. Convert categorical columns to strings
    for col in CATEGORICAL_FEATURES:
        df_dice[col] = df_dice[col].astype(str)

    # 3. Create DiCE Data interface
    continuous_features = [f for f in FEATURE_COLUMNS if f not in CATEGORICAL_FEATURES]
    d = dice_ml.Data(
        dataframe=df_dice,
        continuous_features=continuous_features,
        categorical_features=CATEGORICAL_FEATURES,
        outcome_name='target'
    )

    # 4. Create Model wrapper and explainer
    m = dice_ml.Model(model=rf_model, backend='sklearn')
    dice_explainer = Dice(data_interface=d, model_interface=m, method='random')

    # 5. Build the query instance DataFrame
    sample_df = pd.DataFrame(X_test.iloc[[0]].values, columns=FEATURE_COLUMNS)
    for col in CATEGORICAL_FEATURES:
        sample_df[col] = sample_df[col].astype(float).astype(int).astype(str)

    # 6. Build permitted_range
    permitted_range = {}
    for feat in FEATURE_COLUMNS:
        if feat in CATEGORICAL_FEATURES:
            permitted_range[feat] = sorted(df_dice[feat].unique().tolist())
        else:
            permitted_range[feat] = [float(df_dice[feat].min()), float(df_dice[feat].max())]

    # 7. Generate counterfactuals
    dice_cfs = dice_explainer.generate_counterfactuals(
        sample_df,
        total_CFs=5,
        desired_class='opposite',
        features_to_vary=FEATURE_COLUMNS,
        permitted_range=permitted_range
    )

    # 8. Display counterfactuals
    print("DiCE Counterfactual Explanations for the first test instance:")
    print(dice_cfs.visualize_as_dataframe())

    return dice_cfs

def save_artifacts(models, scaler, le, shap_explainer_rf, shap_explainer_lr, X_train, X_test, y_train, y_test):
    """7. Saves models and explainers to disk."""
    print("\n7. SAVING MODELS AND EXPLAINERS")
    print("-" * 40)

    rf_model, lr_model, nn_model = models['Random Forest'][0], models['Logistic Regression'][0], models['Neural Network'][0]

    # Save models
    joblib.dump(rf_model, 'diabetes_rf_model.pkl')
    joblib.dump(lr_model, 'diabetes_lr_model.pkl')
    joblib.dump(scaler, 'diabetes_scaler.pkl')
    joblib.dump(le, 'diabetes_label_encoder.pkl')
    nn_model.save('diabetes_nn_model.keras')

    # Save SHAP explainers
    with open('shap_explainer_rf.pkl', 'wb') as f:
        pickle.dump(shap_explainer_rf, f)
    with open('shap_explainer_lr.pkl', 'wb') as f:
        pickle.dump(shap_explainer_lr, f)

    # Save feature information
    feature_info = {
        'feature_names': FEATURE_NAMES_HUMAN,
        'feature_columns': FEATURE_COLUMNS
    }
    with open('feature_info.pkl', 'wb') as f:
        pickle.dump(feature_info, f)

    # Save sample data for reference
    sample_data = {
        'X_train_sample': X_train.iloc[:100],
        'X_test_sample': X_test.iloc[:100],
        'y_train_sample': y_train[:100],
        'y_test_sample': y_test[:100]
    }
    with open('sample_data.pkl', 'wb') as f:
        pickle.dump(sample_data, f)

    print("All models and explainers saved successfully!")
    print("\nFiles saved:")
    print("- diabetes_rf_model.pkl")
    print("- diabetes_lr_model.pkl")
    print("- diabetes_nn_model.keras")
    print("- shap_explainer_rf.pkl (RF SHAP Explainer)")
    print("- shap_explainer_lr.pkl (LR SHAP Explainer)")
    print("- diabetes_scaler.pkl (Feature Scaler)")
    print("- diabetes_label_encoder.pkl (Label Encoder)")
    print("- feature_info.pkl (Feature Names)")
    print("- sample_data.pkl (Sample Test Data for Plotting)")

def run_pipeline(data_path=None):
    """Main function to run the entire ML pipeline."""
    print("Starting Comprehensive ML Analysis Pipeline...")
    print("=" * 80)

    # 1. Data Loading and Preprocessing
    df, X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled, scaler, le = \
        load_and_preprocess_data(data_path)

    # 2. Model Training and Evaluation
    models = train_and_evaluate_models(X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled)
    rf_model = models['Random Forest'][0]

    # 3. Explainability Setup
    lime_explainer, shap_explainer_rf, shap_explainer_lr = \
        setup_explainability_tools(X_train, rf_model, models['Logistic Regression'][0], X_train_scaled)

    # 4. Fairness Evaluation
    _ = evaluate_fairness_metrics(X_test, y_test, models)

    # 5. Causality Analysis
    _ = perform_causal_analysis(df)

    # 6. Counterfactual Explanations (DiCE)
    _ = setup_dice_counterfactuals(df, rf_model, X_test)

    # ==============================================================================
    # 7. GENERATE CLINICAL NARRATIVES (NEW STEP)
    # ==============================================================================
    print("\n7. GENERATING CLINICAL NARRATIVES (LOCAL LLM)")
    print("-" * 40)
    
    # Initialize the LLM (Loads model to RTX 4080)
    # This calls the class you just created in llm_layer.py
    try:
        clinical_llm = ClinicalLLM()
        llm_ready = True
    except Exception as e:
        print(f"Skipping LLM step: {e}")
        llm_ready = False
    
    # We will generate narratives for the first 5 patients in the test set
    narrative_records = []
    
    if llm_ready and clinical_llm.model is not None:
        for i in range(5):
            print(f"Processing Patient ID {i}...")
            
            # 1. Get Patient Data
            patient_row = X_test.iloc[i]
            patient_str = patient_row.to_string()
            
            # 2. Get Probability
            # Reshape is needed because predict_proba expects a 2D array
            prob = rf_model.predict_proba(patient_row.values.reshape(1, -1))[0][1]
            
            # 3. Get LIME Features
            # We use the lime_explainer you created in Step 3
            exp = lime_explainer.explain_instance(
                patient_row.values, 
                rf_model.predict_proba, 
                num_features=5
            )
            lime_features = exp.as_list()
            
            # 4. Generate Narrative
            explanation = clinical_llm.generate_explanation(patient_str, prob, lime_features)
            
            # 5. Store for CSV
            narrative_records.append({
                "Patient_ID": i,
                "Risk_Probability": f"{prob:.4f}",
                "Actual_Label": y_test.iloc[i],
                "LIME_Key_Features": str(lime_features),
                "AI_Narrative": explanation
            })
            
            print(f"Narrative generated for Patient {i}.\n")

        # Save the narratives to CSV for your Thesis Review
        output_file = "clinical_narratives_for_review.csv"
        pd.DataFrame(narrative_records).to_csv(output_file, index=False)
        print(f"Successfully saved {len(narrative_records)} narratives to '{output_file}'")
    else:
        print("Skipping narrative generation (LLM failed to load).")

    # ==============================================================================

    # 7. Save Artifacts
    save_artifacts(models, scaler, le, shap_explainer_rf, shap_explainer_lr, X_train, X_test, y_train, y_test)

    print("\n" + "=" * 80)
    print("ML ANALYSIS PIPELINE COMPLETED SUCCESSFULLY!")
    print("=" * 80)


if __name__ == "__main__":
# REPLACE 'YOUR_FILENAME.csv' with the actual name of your file
    dataset_path = r"C:\Users\T2510597\Downloads\xaithesis\Diabetes_Final_Data_V2.csv"
    
    # We pass the path directly to the function
    run_pipeline(data_path=dataset_path)
```

```python
import joblib
import pickle
import pandas as pd
import numpy as np
import shap
import lime
import lime.lime_tabular
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from IPython.display import display
import sys

# Initialize SHAP JavaScript for interactive plot
shap.initjs()

print("Loading saved models and explainers...")
# 1. Load Artifacts
try:
    rf_model = joblib.load('diabetes_rf_model.pkl')
    with open('shap_explainer_rf.pkl', 'rb') as f:
        shap_explainer_rf = pickle.load(f)
    with open('feature_info.pkl', 'rb') as f:
        feature_info = pickle.load(f)
    with open('sample_data.pkl', 'rb') as f:
        sample_data = pickle.load(f)
except FileNotFoundError as e:
    print(f"Error loading files: {e}. Please ensure you ran 'ml_analysis_pipeline.py' first.")

    sys.exit(1)

feature_names = feature_info['feature_names']
X_test = sample_data['X_test_sample']
y_test = sample_data['y_test_sample']
X_train_values = sample_data['X_train_sample'].values # Needed for LIME setup

# Explanation Examples
sample_idx = 45
if sample_idx >= len(X_test):
    sample_idx = 0
sample_instance = X_test.iloc[sample_idx].values
sample_target = y_test.iloc[sample_idx]

print(f"\nExample Instance (Index {sample_idx}):")
print("-" * 30)
for i, (name, value) in enumerate(zip(feature_names, sample_instance)):
    print(f"  {name}: {value}")
print(f"Actual target: {'Diabetes' if sample_target == 1 else 'No Diabetes'}")


## 1. LIME Explanation
print("\n[1] Generating LIME Explanation (Top 5 features)...")

# Re-initialize LIME explainer with loaded training data
lime_explainer = lime.lime_tabular.LimeTabularExplainer(
    X_train_values,
    feature_names=feature_names,
    class_names=['No Diabetes', 'Diabetes'],
    mode='classification'
)

# Function to wrap prediction for LIME
def rf_predict_proba(x):

    if x.ndim == 1:
        x = x.reshape(1, -1)
    return rf_model.predict_proba(x)

lime_exp = lime_explainer.explain_instance(sample_instance, rf_predict_proba, num_features=5)

print("\nLIME Explanation (Top 5 features):")
for feature, importance in lime_exp.as_list():
    print(f"  {feature}: {importance:.4f}")


## 2. SHAP Plots
print("\n[2] Generating SHAP Plots (using Random Forest)...")

# Calculate SHAP values for the sample instance
sample_instance_shap = sample_instance.astype(float).reshape(1, -1)
shap_values_sample = shap_explainer_rf.shap_values(sample_instance_shap)

if isinstance(shap_values_sample, list) and len(shap_values_sample) == 2:
    shap_values_for_class = shap_values_sample[1][0]
    base_value = shap_explainer_rf.expected_value[1]

elif isinstance(shap_values_sample, np.ndarray) and shap_values_sample.ndim == 3:
    shap_values_for_class = shap_values_sample[0, :, 1]
    base_value = shap_explainer_rf.expected_value[1]

else:
    if len(shap_explainer_rf.expected_value) > 1:
        base_value = shap_explainer_rf.expected_value[1]
    else:
        base_value = shap_explainer_rf.expected_value
    shap_values_for_class = shap_values_sample[0]

base_value_float = float(base_value)

# a) SHAP Local Force Plot (Interactive)
print("\nGenerating SHAP Local Force Plot (Interactive)...")
force_plot = shap.force_plot(
    base_value_float,
    shap_values_for_class,
    X_test.iloc[sample_idx],
    feature_names=feature_names,
    matplotlib=False
)
display(force_plot)

# b) SHAP Local Bar Plot
print("\nGenerating SHAP Local Bar Plot (single instance)...")
shap.summary_plot(
    shap_values_for_class.reshape(1, -1),
    X_test.iloc[sample_idx:sample_idx+1],
    feature_names=feature_names,
    plot_type='bar',
    show=False
)
plt.title("SHAP Local Bar Plot")
plt.tight_layout()
plt.show()

shap_values_test_sample = shap_explainer_rf.shap_values(X_test)

# c) SHAP Global Summary Plot
print("\nGenerating SHAP Global Summary Plot (first 100 instances)...")

if isinstance(shap_values_test_sample, list) and len(shap_values_test_sample) == 2:
    shap_values_for_summary = shap_values_test_sample[1]

elif isinstance(shap_values_test_sample, np.ndarray) and shap_values_test_sample.ndim == 3:
    shap_values_for_summary = shap_values_test_sample[:, :, 1]

else:
    shap_values_for_summary = shap_values_test_sample

shap.summary_plot(
    shap_values_for_summary,
    X_test,
    feature_names=feature_names,
    show=False
)

plt.title("SHAP Global Summary Plot (Diabetes Class)")
plt.tight_layout()
plt.show()

# -------------------------------
# d) SHAP Scatter/Dependence Plot
# -------------------------------
print("\nGenerating SHAP Scatter/Dependence Plot (Glucose vs. BMI)...")

primary_feature_name = 'Glucose'
interaction_feature_name = 'BMI'
plot_dependencies = True
# Get the index of the features using the feature names list
try:
    glucose_idx = feature_names.index(primary_feature_name)
    bmi_idx = feature_names.index(interaction_feature_name)
except ValueError:
    print(f"Error: Could not find '{primary_feature_name}' or '{interaction_feature_name}' in feature names. Skipping plot.")
    plot_dependencies = pickle.FALSE

if plot_dependencies:

    # Generate the Dependence Plot
    shap.dependence_plot(
        ind=glucose_idx,
        shap_values=shap_values_for_summary,
        features=X_test,
        feature_names=feature_names,
        interaction_index=bmi_idx,
        show=False
    )
    plt.title(f"SHAP Dependence Plot: {primary_feature_name} vs. {interaction_feature_name}")
    plt.tight_layout()
    plt.show()

    print("SHAP Scatter/Dependence Plot generated successfully.")

print("\nExplanation visualizations completed.")
```
