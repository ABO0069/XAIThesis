```python
# ml_analysis_pipeline.py
# Comprehensive ML Pipeline for Diabetes Prediction with Explainable AI
# Features: Class-balanced training, XGBoost, SHAP/LIME/DiCE, Fairness, Causality, LLM narratives

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import pickle
import joblib
import sys

# Core ML & preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
import xgboost as xgb

# Deep learning
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Explainability
import lime
import lime.lime_tabular
import shap
import dice_ml
from dice_ml import Dice

# Fairness & bias
from aif360.datasets import BinaryLabelDataset

# Causality
from dowhy import CausalModel

# LLM Layer (required)
# from llm_layer import ClinicalLLM  # Ensure this file exists in your directory

# Colab-specific utility
try:
    from google.colab import files
    COLAB_ENV = True
except ImportError:
    COLAB_ENV = False

warnings.filterwarnings('ignore')
np.random.seed(42)

# CONSTANTS
RANDOM_STATE = 42
TEST_SIZE = 0.2  # 80% Training, 20% Testing

# Define all feature columns
FEATURE_COLUMNS = [
    'age', 'gender_encoded', 'pulse_rate', 'systolic_bp', 'diastolic_bp',
    'glucose', 'height', 'weight', 'bmi', 'family_diabetes', 'hypertensive',
    'family_hypertension', 'cardiovascular_disease', 'stroke'
]

# Feature names for interpretability
FEATURE_NAMES_HUMAN = [
    'Age', 'Gender', 'Pulse Rate', 'Systolic BP', 'Diastolic BP',
    'Glucose', 'Height', 'Weight', 'BMI', 'Family Diabetes', 'Hypertensive',
    'Family Hypertension', 'Cardiovascular Disease', 'Stroke'
]

# Categorical features for DiCE/Fairness
CATEGORICAL_FEATURES = [
    'gender_encoded','family_diabetes','hypertensive',
    'family_hypertension','cardiovascular_disease','stroke'
]

# Medical Feature Taxonomy for Counterfactuals
IMMUTABLE_FEATURES = [
    'age', 'gender_encoded', 'family_diabetes', 
    'family_hypertension', 'height'
]

MODIFIABLE_FEATURES = [
    'pulse_rate', 'systolic_bp', 'diastolic_bp', 'glucose', 
    'weight', 'bmi', 'hypertensive', 'cardiovascular_disease', 'stroke'
]

# ---------------------------------

def load_and_preprocess_data(data_path=None):
    """1. Loads, encodes, splits, and scales the data."""
    print("\n1. LOADING AND PREPROCESSING DATA")
    print("-" * 40)
    df = None

    if COLAB_ENV and data_path is None:
        print("Colab Environment Detected. Opening file upload dialog...")
        uploaded = files.upload()
        if not uploaded:
            raise FileNotFoundError("No file uploaded in Colab session.")
        fn = next(iter(uploaded.keys()))
        df = pd.read_csv(fn)
        print(f"Uploaded file: {fn}")
    elif data_path:
        print(f"Reading data from provided path: {data_path}")
        df = pd.read_csv(data_path)
    else:
        raise ValueError("Cannot load data. Please provide a `data_path` parameter or run in a Colab environment to prompt upload.")

    if df is not None:
        print("\nFirst 5 rows of the loaded dataset:")
        print(df.head())

    # Encode categorical and target variables
    le = LabelEncoder()
    df['gender_encoded'] = le.fit_transform(df['gender'])
    df['target'] = le.fit_transform(df['diabetic'])
    X = df[FEATURE_COLUMNS]
    y = df['target']

    print(f"\nOriginal class distribution: {np.bincount(y)}")
    print(f"   Non-Diabetic: {np.sum(y == 0)} ({np.sum(y == 0) / len(y) * 100:.1f}%)")
    print(f"   Diabetic: {np.sum(y == 1)} ({np.sum(y == 1) / len(y) * 100:.1f}%)")
    print("\n Using class balancing via model parameters to handle imbalance")

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=TEST_SIZE, 
        random_state=RANDOM_STATE, 
        stratify=y
    )

    print(f"\nTrain set distribution: {np.bincount(y_train)}")
    print(f"Test set distribution: {np.bincount(y_test)}")

    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    return df, X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled, scaler, le

def train_and_evaluate_models(X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled, scaler):
    """2. Trains Random Forest, XGBoost, Logistic Regression, and Neural Network models."""
    print("\n2. TRAINING MODELS WITH CLASS BALANCING (FULL TRAINING SET)")
    print("-" * 40)

    # Random Forest with class balancing
    print("Training Random Forest...")
    rf_model = RandomForestClassifier(
        n_estimators=200,           # Increased from 100
        max_depth=15,               # Increased from 10
        min_samples_split=10,       # Prevent overfitting
        min_samples_leaf=5,
        class_weight='balanced',    #  CRITICAL FIX
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
    rf_model.fit(X_train, y_train)
    rf_predictions = rf_model.predict(X_test)
    rf_probabilities = rf_model.predict_proba(X_test)[:, 1]

    # XGBoost Model
    print("Training XGBoost...")
    
    # Calculate scale_pos_weight for class imbalance
    scale_pos_weight = np.sum(y_train == 0) / np.sum(y_train == 1)
    
    xgb_model = xgb.XGBClassifier(
        n_estimators=200,
        max_depth=6,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        scale_pos_weight=scale_pos_weight,  #  Handles imbalance
        random_state=RANDOM_STATE,
        eval_metric='logloss',
        use_label_encoder=False
    )
    xgb_model.fit(X_train, y_train)
    xgb_predictions = xgb_model.predict(X_test)
    xgb_probabilities = xgb_model.predict_proba(X_test)[:, 1]

    # Logistic Regression Model with class_weight
    print("Training Logistic Regression...")
    lr_model = LogisticRegression(
        random_state=RANDOM_STATE, 
        max_iter=1000,
        class_weight='balanced'
    )
    lr_model.fit(X_train_scaled, y_train)
    lr_predictions = lr_model.predict(X_test_scaled)
    lr_probabilities = lr_model.predict_proba(X_test_scaled)[:, 1]

    # Neural Network Model with class weighting
    print("Training Neural Network...")
    
    # Calculate class weights for Keras
    class_weights = {
        0: len(y_train) / (2 * np.sum(y_train == 0)),
        1: len(y_train) / (2 * np.sum(y_train == 1))
    }
    
    nn_model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(16, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    nn_model.fit(
        X_train_scaled, y_train,
        validation_split=0.2, 
        epochs=100, 
        batch_size=32,
        callbacks=[early_stopping], 
        verbose=0,
        class_weight=class_weights
    )
    nn_probabilities = nn_model.predict(X_test_scaled).flatten()
    nn_predictions = (nn_probabilities > 0.5).astype(int)

    models = {
        'Random Forest': (rf_model, rf_predictions, rf_probabilities),
        'XGBoost': (xgb_model, xgb_predictions, xgb_probabilities),
        'Logistic Regression': (lr_model, lr_predictions, lr_probabilities),
        'Neural Network': (nn_model, nn_predictions, nn_probabilities)
    }

    print("\n" + "=" * 75)
    print("MODEL PERFORMANCE COMPARISON")
    print("=" * 75)
    print(f"{'Model':20} {'Acc':>6} {'Precision':>10} {'Recall':>8} {'F1':>6} {'AUC':>6}")
    print("-" * 75)
    for name, (_, pred, prob) in models.items():
        acc = accuracy_score(y_test, pred)
        prec = precision_score(y_test, pred, zero_division=0)
        rec = recall_score(y_test, pred, zero_division=0)
        f1 = f1_score(y_test, pred)
        auc = roc_auc_score(y_test, prob)
        print(f"{name:20} {acc:>6.4f} {prec:>10.4f} {rec:>8.4f} {f1:>6.4f} {auc:>6.4f}")
    print("=" * 75)

    # Plot ROC Curves
    plt.figure(figsize=(10, 7))
    for name, (_, _, prob) in models.items():
        fpr, tpr, _ = roc_curve(y_test, prob)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc_score(y_test, prob):.4f})', linewidth=2)
    plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier', linewidth=2)
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate (Recall)', fontsize=12)
    plt.title('ROC Curves - All Models', fontsize=14, fontweight='bold')
    plt.legend(loc='lower right', fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('roc_curves.png', dpi=300)
    plt.show()

    # Test extreme case
    print("\n" + "=" * 75)
    print(" SANITY CHECK: Testing Extreme Diabetic Patient")
    print("=" * 75)
    extreme_patient = pd.DataFrame([{
        'age': 60,
        'gender_encoded': 1,
        'pulse_rate': 95,
        'systolic_bp': 180,
        'diastolic_bp': 110,
        'glucose': 18.0,  # Very high glucose
        'height': 170,
        'weight': 95,
        'bmi': 32.9,
        'family_diabetes': 1,
        'hypertensive': 1,
        'family_hypertension': 1,
        'cardiovascular_disease': 1,
        'stroke': 0
    }])
    
    extreme_patient_scaled = scaler.transform(extreme_patient)
    
    print("Patient Profile: Age=60, Glucose=18.0, BP=180/110, BMI=32.9")
    print("\nModel Predictions:")
    print(f"   Random Forest:  {rf_model.predict_proba(extreme_patient)[0][1]:.1%} risk")
    print(f"   XGBoost:        {xgb_model.predict_proba(extreme_patient)[0][1]:.1%} risk")
    print(f"   Neural Network: {nn_model.predict(extreme_patient_scaled)[0][0]:.1%} risk")
    print("\n If these are >70%, the fix worked!")
    print("=" * 75)

    return models

def setup_explainability_tools(X_train, rf_model, lr_model, X_train_scaled):
    """3. Initializes LIME and SHAP explainers."""
    print("\n3. SETTING UP EXPLAINABILITY TOOLS")
    print("-" * 40)

    # LIME Setup
    lime_explainer = lime.lime_tabular.LimeTabularExplainer(
        X_train.values,
        feature_names=FEATURE_NAMES_HUMAN,
        class_names=['No Diabetes', 'Diabetes'],
        mode='classification',
        discretize_continuous=True
    )

    # SHAP Setup
    # Use a background sample of 500 instances to speed up SHAP
    X_background_sample = X_train.sample(n=500, random_state=RANDOM_STATE).values.astype(float)
    shap_explainer_rf = shap.TreeExplainer(rf_model, X_background_sample)
    shap_explainer_lr = shap.LinearExplainer(lr_model, X_train_scaled)

    print(" LIME and SHAP explainers initialized successfully!")
    return lime_explainer, shap_explainer_rf, shap_explainer_lr

def evaluate_fairness_metrics(X_test, y_test, models):
    """4. Evaluates fairness metrics for all models."""
    print("\n4. FAIRNESS EVALUATION")
    print("-" * 40)
    
    def calculate_fairness(X_test_df, y_test_series, predictions, sensitive_features=['gender_encoded', 'age']):
        results = {}
        for feature in sensitive_features:
            feature_values = X_test_df[feature].values.copy()
            if feature == 'age':
                bins = [0, 35, 60, 120]
                labels = ['Young', 'Middle', 'Old']
                groups = pd.cut(feature_values, bins=bins, labels=labels, include_lowest=True)
            else:
                groups = feature_values

            feature_result = {}
            for group in np.unique(groups):
                mask = groups == group
                if np.sum(mask) > 0:
                    group_accuracy = accuracy_score(y_test_series[mask], predictions[mask])
                    group_precision = precision_score(y_test_series[mask], predictions[mask], zero_division=0)
                    group_recall = recall_score(y_test_series[mask], predictions[mask], zero_division=0)

                    feature_result[group] = {
                        'accuracy': group_accuracy,
                        'precision': group_precision,
                        'recall': group_recall,
                        'count': np.sum(mask)
                    }

            if len(feature_result) > 1:
                first_group = list(feature_result.keys())[0]
                last_group = list(feature_result.keys())[-1]

                disparate_impact = feature_result[first_group]['accuracy'] / feature_result[last_group]['accuracy'] if feature_result[last_group]['accuracy'] > 0 else 0.0
                feature_result['disparate_impact'] = disparate_impact
                feature_result['fairness_satisfied'] = 0.8 <= disparate_impact <= 1.25

            results[feature] = feature_result
        return results

    X_test_df_reset = X_test.reset_index(drop=True)
    y_test_series_reset = y_test.reset_index(drop=True)

    # Create output string for both console and file
    output_lines = []
    output_lines.append("\nFairness Analysis Across Demographic Groups:")
    output_lines.append("=" * 80)
    
    fairness_results_all_models = {}
    
    for name, (_, pred, _) in models.items():
        fairness_results = calculate_fairness(X_test_df_reset, y_test_series_reset, pred)
        fairness_results_all_models[name] = fairness_results
        
        output_lines.append(f"\n{name}:")
        output_lines.append("-" * 80)
        
        for feature, values in fairness_results.items():
            if feature == 'gender_encoded':
                feature_display = "Gender"
            else:
                feature_display = feature.title()
            
            output_lines.append(f"\n   {feature_display} Analysis:")
            
            # Print group metrics
            for key, value in values.items():
                if isinstance(value, dict):
                    group_name = f"Group {key}" if feature == 'gender_encoded' else key
                    line = f"    {group_name:12} | Acc: {value['accuracy']:.3f} | Prec: {value['precision']:.3f} | Recall: {value['recall']:.3f} | n={value['count']:3d}"
                    output_lines.append(line)
            
            # Print fairness metrics
            if 'disparate_impact' in values:
                di_ratio = values['disparate_impact']
                fairness_status = " FAIR" if values['fairness_satisfied'] else " UNFAIR"
                output_lines.append(f"    Disparate Impact Ratio: {di_ratio:.3f} [{fairness_status}]")
                output_lines.append(f"    (Fair range: 0.80 - 1.25)")
    
    # Summary table
    output_lines.append("\n" + "=" * 80)
    output_lines.append("FAIRNESS SUMMARY")
    output_lines.append("=" * 80)
    output_lines.append(f"{'Model':<20} {'Gender Fair?':<15} {'Age Fair?':<15}")
    output_lines.append("-" * 80)
    
    for name, results in fairness_results_all_models.items():
        gender_fair = " Yes" if results.get('gender_encoded', {}).get('fairness_satisfied', False) else " No"
        age_fair = " Yes" if results.get('age', {}).get('fairness_satisfied', False) else " No"
        output_lines.append(f"{name:<20} {gender_fair:<15} {age_fair:<15}")
    
    output_lines.append("=" * 80)
    
    # Print to console
    for line in output_lines:
        print(line)
    
    # Save to file with UTF-8 encoding to handle special characters
    fairness_output_file = "fairness_evaluation_results.txt"
    with open(fairness_output_file, 'w', encoding='utf-8') as f:
        f.write('\n'.join(output_lines))
    
    print(f"\n Fairness results saved to '{fairness_output_file}'")
    
    return fairness_results_all_models

def perform_causal_analysis(df):
    """5. Performs a sample causal analysis using DoWhy."""
    print("\n5. CAUSALITY ANALYSIS")
    print("-" * 40)

    def analyze(df, treatment_var, outcome_var, confounders):
        try:
            df_for_causal = df.copy()
            
            # Ensure target column exists
            if 'target' not in df_for_causal.columns:
                if 'diabetic' in df_for_causal.columns:
                    le = LabelEncoder()
                    df_for_causal['target'] = le.fit_transform(df_for_causal['diabetic'])
                else:
                    print(" ERROR: Neither 'target' nor 'diabetic' column found!")
                    return None
            
            # Create treatment variable
            if treatment_var not in df_for_causal.columns or treatment_var != 'high_glucose':
                df_for_causal['high_glucose'] = (df_for_causal['glucose'] > df_for_causal['glucose'].median()).astype(int)

            # Verify confounders exist
            missing_confounders = [c for c in confounders if c not in df_for_causal.columns]
            if missing_confounders:
                print(f" ERROR: Missing confounders: {missing_confounders}")
                return None

            print(f" Treatment variable '{treatment_var}' created (median split)")
            print(f" Outcome variable: '{outcome_var}'")
            print(f" Confounders: {confounders}")
            
            causal_model = CausalModel(
                data=df_for_causal,
                treatment=treatment_var,
                outcome=outcome_var,
                confounders=confounders
            )
            
            print("\n Identifying causal estimand...")
            identified_estimand = causal_model.identify_effect()
            
            print(" Estimating causal effect using backdoor adjustment...")
            causal_estimate = causal_model.estimate_effect(
                identified_estimand,
                method_name="backdoor.linear_regression"
            )
            
            print("\n" + "=" * 70)
            print("CAUSAL ANALYSIS RESULTS")
            print("=" * 70)
            print(f"Causal effect of {treatment_var} on {outcome_var}: {causal_estimate.value:.4f}")
            
            conf_intervals = causal_estimate.get_confidence_intervals()
            lower_bound = float(conf_intervals[0, 0])
            upper_bound = float(conf_intervals[0, 1])
            print(f"95% Confidence Interval: [{lower_bound:.4f}, {upper_bound:.4f}]")
            
            # Interpretation
            print("\nINTERPRETATION:")
            if causal_estimate.value > 0:
                print(f" High glucose levels INCREASE diabetes risk by {causal_estimate.value:.1%}")
                print(f"  (After controlling for: {', '.join(confounders)})")
            else:
                print(f" No significant causal effect detected")
            print("=" * 70)
            
            return causal_estimate

        except Exception as e:
            print(f"\n CAUSAL ANALYSIS FAILED: {str(e)}")
            import traceback
            print(traceback.format_exc())
            return None

    confounders = ['age', 'bmi', 'systolic_bp', 'family_diabetes']
    causal_result = analyze(df, treatment_var='high_glucose', outcome_var='target', confounders=confounders)
    
    if causal_result is None:
        print("\n  Causal analysis could not be completed. Check data requirements.")
    
    return causal_result

def setup_dice_counterfactuals(df, rf_model, X_test):
    """6. Sets up medically-constrained DiCE counterfactual explanations."""
    print("\n6. MEDICALLY-CONSTRAINED COUNTERFACTUAL EXPLANATIONS")
    print("-" * 40)

    # Prepare df_dice
    df_dice = df.drop(columns=['gender','diabetic'], errors='ignore')
    df_dice = df_dice[FEATURE_COLUMNS + ['target']]

    # Convert categorical columns to strings
    for col in CATEGORICAL_FEATURES:
        df_dice[col] = df_dice[col].astype(str)

    # Create DiCE Data interface
    continuous_features = [f for f in FEATURE_COLUMNS if f not in CATEGORICAL_FEATURES]
    d = dice_ml.Data(
        dataframe=df_dice,
        continuous_features=continuous_features,
        categorical_features=CATEGORICAL_FEATURES,
        outcome_name='target'
    )

    # Create Model wrapper and explainer
    m = dice_ml.Model(model=rf_model, backend='sklearn')
    dice_explainer = Dice(data_interface=d, model_interface=m, method='random')

    # Find a high-risk diabetic patient for demonstration
    rf_probabilities = rf_model.predict_proba(X_test)[:, 1]
    diabetic_patients = np.where(rf_probabilities > 0.6)[0]
    
    if len(diabetic_patients) == 0:
        print("  No high-risk patients found in test set for counterfactual demo")
        return None
    
    patient_idx = diabetic_patients[0]
    sample_df = pd.DataFrame(X_test.iloc[[patient_idx]].values, columns=FEATURE_COLUMNS)
    
    # Convert categorical features to strings
    for col in CATEGORICAL_FEATURES:
        if col in sample_df.columns:
            sample_df[col] = sample_df[col].astype(float).astype(int).astype(str)

    # Build medically-constrained permitted_range
    permitted_range = {}
    
    for feat in FEATURE_COLUMNS:
        if feat in CATEGORICAL_FEATURES:
            current_value = str(sample_df[feat].iloc[0])
        else:
            current_value = float(sample_df[feat].iloc[0])
        
        if feat in IMMUTABLE_FEATURES:
            # IMMUTABLE: Lock to current value
            if feat in CATEGORICAL_FEATURES:
                permitted_range[feat] = [current_value]
            else:
                permitted_range[feat] = [current_value, current_value]
        
        elif feat in CATEGORICAL_FEATURES:
            # Categorical modifiable features
            unique_vals = sorted(df_dice[feat].unique().tolist())
            
            # For binary health conditions, only allow improvement (1 -> 0)
            if feat in ['hypertensive', 'cardiovascular_disease', 'stroke']:
                if current_value == '1':
                    permitted_range[feat] = ['0', '1']
                else:
                    permitted_range[feat] = ['0']
            else:
                permitted_range[feat] = unique_vals
        
        else:
            # Continuous modifiable features: Allow 50% change (relaxed from 30%)
            data_min = float(df_dice[feat].min())
            data_max = float(df_dice[feat].max())
            
            # Wider range for DiCE to find solutions
            max_change = abs(current_value * 0.50)  # 50% change allowed
            lower_bound = max(data_min, current_value - max_change)
            upper_bound = min(data_max, current_value + max_change)
            
            permitted_range[feat] = [float(lower_bound), float(upper_bound)]

    # Generate counterfactuals with relaxed constraints
    print(f"\nGenerating counterfactuals for Patient {patient_idx} (Risk: {rf_probabilities[patient_idx]:.1%})...")
    try:
        dice_cfs = dice_explainer.generate_counterfactuals(
            sample_df,
            total_CFs=3,  # Reduced from 5 for better success rate
            desired_class='opposite',
            features_to_vary=MODIFIABLE_FEATURES,
            permitted_range=permitted_range
        )

        # Display counterfactuals
        print("\n Medically-Constrained Counterfactual Explanations:")
        print("="*70)
        cf_df = dice_cfs.visualize_as_dataframe()
        print(cf_df)
        
        print("\n" + "="*70)
        print("CONSTRAINT COMPLIANCE:")
        print("    Age, Gender, Family History: LOCKED (unchanged)")
        print("    Modifiable features: Limited to 50% realistic change")
        print("    Health conditions: Only improvement allowed")
        print("="*70)
        
        return dice_cfs
        
    except Exception as e:
        print(f"  DiCE generation failed: {e}")
        print("Note: Counterfactuals may not exist within strict medical constraints.")
        print("Continuing with other analyses...")
        return None

def save_artifacts(models, scaler, le, shap_explainer_rf, shap_explainer_lr, X_train, X_test, y_train, y_test):
    """7. Saves models and explainers to disk."""
    print("\n8. SAVING MODELS AND EXPLAINERS")
    print("-" * 40)

    rf_model = models['Random Forest'][0]
    xgb_model = models['XGBoost'][0]
    lr_model = models['Logistic Regression'][0]
    nn_model = models['Neural Network'][0]

    # Save models
    joblib.dump(rf_model, 'diabetes_rf_model.pkl')
    joblib.dump(xgb_model, 'diabetes_xgb_model.pkl')
    joblib.dump(lr_model, 'diabetes_lr_model.pkl')
    joblib.dump(scaler, 'diabetes_scaler.pkl')
    joblib.dump(le, 'diabetes_label_encoder.pkl')
    nn_model.save('diabetes_nn_model.keras')

    # Save SHAP explainers
    with open('shap_explainer_rf.pkl', 'wb') as f:
        pickle.dump(shap_explainer_rf, f)
    with open('shap_explainer_lr.pkl', 'wb') as f:
        pickle.dump(shap_explainer_lr, f)

    # Save feature information
    feature_info = {
        'feature_names': FEATURE_NAMES_HUMAN,
        'feature_columns': FEATURE_COLUMNS
    }
    with open('feature_info.pkl', 'wb') as f:
        pickle.dump(feature_info, f)

    # ================================================================
    # UPDATED: SAVE FULL TEST SET (No more 100 sample limit)
    # ================================================================
    print(f"Saving {len(X_test)} test samples for visualization (Full Test Set)...")
    
    # We save the ENTIRE test set so the visualization tool can access everything.
    # We also keep the 'X_train_sample' small (500) because it's only used for SHAP background.
    sample_data = {
        'X_train_sample': X_train.sample(n=min(len(X_train), 500), random_state=42),
        'X_test_sample': X_test,  # <--- SAVING FULL TEST SET
        'y_train_sample': y_train.iloc[:min(len(y_train), 500)],
        'y_test_sample': y_test   # <--- SAVING FULL TARGETS
    }
    with open('sample_data.pkl', 'wb') as f:
        pickle.dump(sample_data, f)

    print(" All models and explainers saved successfully!")
    print("\nFiles saved:")
    print("- diabetes_rf_model.pkl (Random Forest)")
    print("- diabetes_xgb_model.pkl (XGBoost)  NEW!")
    print("- diabetes_lr_model.pkl (Logistic Regression)")
    print("- diabetes_nn_model.keras (Neural Network)")
    print("- shap_explainer_rf.pkl (RF SHAP Explainer)")
    print("- shap_explainer_lr.pkl (LR SHAP Explainer)")
    print("- diabetes_scaler.pkl (Feature Scaler)")
    print("- diabetes_label_encoder.pkl (Label Encoder)")
    print("- feature_info.pkl (Feature Names)")
    print("- sample_data.pkl (Full Test Data for Visualization)")

def run_pipeline(data_path=None):
    """Main function to run the entire ML pipeline."""
    print("=" * 80)
    print("COMPREHENSIVE ML ANALYSIS PIPELINE")
    print("=" * 80)
    print("\nKEY FEATURES:")
    print("    Class-balanced training (handles imbalanced dataset)")
    print("    4 models: Random Forest, XGBoost, Logistic Regression, Neural Network")
    print("    XAI: SHAP, LIME, DiCE counterfactuals")
    print("    Fairness & Causality analysis")
    print("    LLM-generated clinical narratives")
    print("\n" + "=" * 80)

    # 1. Data Loading and Preprocessing
    df, X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled, scaler, le = \
        load_and_preprocess_data(data_path)

    # 2. Model Training and Evaluation
    models = train_and_evaluate_models(X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled, scaler)
    
    # Select best model (you can change this based on results)
    rf_model = models['Random Forest'][0]
    xgb_model = models['XGBoost'][0]

    # 3. Explainability Setup
    lime_explainer, shap_explainer_rf, shap_explainer_lr = \
        setup_explainability_tools(X_train, rf_model, models['Logistic Regression'][0], X_train_scaled)

    # 4. Fairness Evaluation
    _ = evaluate_fairness_metrics(X_test, y_test, models)

    # 5. Causality Analysis
    _ = perform_causal_analysis(df)

    # 6. Counterfactual Explanations
    _ = setup_dice_counterfactuals(df, rf_model, X_test)

    # 7. GENERATE CLINICAL NARRATIVES WITH COUNTERFACTUALS (LLM)
    print("\n7. GENERATING CLINICAL NARRATIVES WITH INTERVENTIONS (LOCAL LLM)")
    print("-" * 40)
    
    # Initialize the LLM
    try:
        from llm_layer import ClinicalLLM
        clinical_llm = ClinicalLLM()
        llm_available = True
    except ImportError:
        print(" llm_layer.py not found. Skipping LLM generation.")
        llm_available = False
    
    if llm_available:
        # Prepare DiCE for counterfactual generation
        df_dice = df.drop(columns=['gender','diabetic'], errors='ignore')
        df_dice = df_dice[FEATURE_COLUMNS + ['target']]
        for col in CATEGORICAL_FEATURES:
            df_dice[col] = df_dice[col].astype(str)
        
        continuous_features = [f for f in FEATURE_COLUMNS if f not in CATEGORICAL_FEATURES]
        d = dice_ml.Data(
            dataframe=df_dice,
            continuous_features=continuous_features,
            categorical_features=CATEGORICAL_FEATURES,
            outcome_name='target'
        )
        m = dice_ml.Model(model=rf_model, backend='sklearn')
        dice_explainer = Dice(data_interface=d, model_interface=m, method='random')
        
        # Generate narratives for the first 5 patients
        narrative_records = []
        
        for i in range(5):
            print(f"Processing Patient ID {i}...")
            
            # Get Patient Data
            patient_row = X_test.iloc[i]
            patient_str = patient_row.to_string()
            
            # Get Probability (using best model - choose RF or XGB)
            prob = rf_model.predict_proba(patient_row.values.reshape(1, -1))[0][1]
            
            # Get LIME Features
            exp = lime_explainer.explain_instance(
                patient_row.values, 
                rf_model.predict_proba, 
                num_features=5,
                top_labels=2
            )
            pred_class = int(rf_model.predict([patient_row.values])[0])
            
            try:
                lime_features = exp.as_list(label=pred_class)
            except KeyError:
                lime_features = exp.as_list(label=1)
            
            # Generate counterfactuals for THIS patient (only if diabetic)
            counterfactual_interventions = []
            if prob > 0.5:
                try:
                    query_df = pd.DataFrame([patient_row.values], columns=FEATURE_COLUMNS)
                    for col in CATEGORICAL_FEATURES:
                        if col in query_df.columns:
                            query_df[col] = query_df[col].astype(float).astype(int).astype(str)
                    
                    permitted_range = {}
                    for feat in FEATURE_COLUMNS:
                        if feat in CATEGORICAL_FEATURES:
                            current_value = str(query_df[feat].iloc[0])
                        else:
                            current_value = float(query_df[feat].iloc[0])
                        
                        if feat in IMMUTABLE_FEATURES:
                            if feat in CATEGORICAL_FEATURES:
                                permitted_range[feat] = [current_value]
                            else:
                                permitted_range[feat] = [current_value, current_value]
                        elif feat in CATEGORICAL_FEATURES:
                            unique_vals = sorted(df_dice[feat].unique().tolist())
                            if feat in ['hypertensive', 'cardiovascular_disease', 'stroke']:
                                if current_value == '1':
                                    permitted_range[feat] = ['0', '1']
                                else:
                                    permitted_range[feat] = ['0']
                            else:
                                permitted_range[feat] = unique_vals
                        else:
                            data_min = float(df_dice[feat].min())
                            data_max = float(df_dice[feat].max())
                            max_change = abs(current_value * 0.30)
                            lower_bound = max(data_min, current_value - max_change)
                            upper_bound = min(data_max, current_value + max_change)
                            permitted_range[feat] = [float(lower_bound), float(upper_bound)]
                    
                    dice_cfs = dice_explainer.generate_counterfactuals(
                        query_df,
                        total_CFs=3,
                        desired_class='opposite',
                        features_to_vary=MODIFIABLE_FEATURES,
                        permitted_range=permitted_range
                    )
                    
                    cf_df = dice_cfs.cf_examples_list[0].final_cfs_df
                    
                    if cf_df is not None and len(cf_df) > 0:
                        cf_row = cf_df.iloc[0]
                        
                        for feat in MODIFIABLE_FEATURES:
                            if feat in cf_row.index and feat in patient_row.index:
                                if feat in CATEGORICAL_FEATURES:
                                    orig_val = str(int(float(patient_row[feat])))
                                    cf_val = str(int(float(cf_row[feat])))
                                else:
                                    orig_val = float(patient_row[feat])
                                    cf_val = float(cf_row[feat])
                                
                                if orig_val != cf_val:
                                    if feat not in CATEGORICAL_FEATURES:
                                        change_pct = ((cf_val - orig_val) / orig_val * 100) if orig_val != 0 else 0
                                        
                                        intervention_map = {
                                            'glucose': ('Glucose Management', ['Diet modifications', 'Oral medication (Metformin)']),
                                            'systolic_bp': ('Blood Pressure Control', ['ACE inhibitors', 'DASH diet']),
                                            'diastolic_bp': ('Blood Pressure Control', ['ARBs', 'Sodium restriction']),
                                            'weight': ('Weight Management', ['Caloric deficit', 'Exercise 150min/week']),
                                            'bmi': ('Weight Management', ['Diet plan', 'Physical activity']),
                                            'pulse_rate': ('Cardiovascular Fitness', ['Aerobic exercise', 'Gradual progression'])
                                        }
                                        
                                        intervention_info = intervention_map.get(feat, ('Lifestyle Modification', ['Health improvements']))
                                        
                                        counterfactual_interventions.append({
                                            'feature': FEATURE_NAMES_HUMAN[FEATURE_COLUMNS.index(feat)],
                                            'current': orig_val,
                                            'target': cf_val,
                                            'change_pct': change_pct,
                                            'intervention': intervention_info[0],
                                            'methods': intervention_info[1],
                                            'timeframe': '3-6 months' if abs(change_pct) < 20 else '6-12 months'
                                        })
                    
                except Exception as e:
                    print(f"   Could not generate counterfactuals: {e}")
            
            # Generate narratives
            if hasattr(clinical_llm, 'generate_patient_narrative'):
                patient_narrative = clinical_llm.generate_patient_narrative(
                    patient_str, prob, lime_features, counterfactual_interventions
                )
                doctor_report = clinical_llm.generate_doctor_report(
                    patient_str, prob, lime_features, counterfactual_interventions
                )
            else:
                patient_narrative = clinical_llm.generate_explanation(patient_str, prob, lime_features)
                doctor_report = patient_narrative
            
            narrative_records.append({
                "Patient_ID": i,
                "Risk_Probability": f"{prob:.4f}",
                "Actual_Label": y_test.iloc[i],
                "LIME_Key_Features": str(lime_features),
                "Interventions": str(counterfactual_interventions) if counterfactual_interventions else "Low risk - no interventions needed",
                "Patient_Narrative": patient_narrative,
                "Doctor_Report": doctor_report
            })
            
            print(f"Narrative generated for Patient {i}.\n")

        # Save the narratives to CSV
        output_file = "clinical_narratives_with_interventions.csv"
        pd.DataFrame(narrative_records).to_csv(output_file, index=False)
        print(f" Successfully saved {len(narrative_records)} narratives to '{output_file}'")

    # 8. Save Artifacts
    save_artifacts(models, scaler, le, shap_explainer_rf, shap_explainer_lr, X_train, X_test, y_train, y_test)

    print("\n" + "=" * 80)
    print(" ML ANALYSIS PIPELINE COMPLETED SUCCESSFULLY!")
    print("=" * 80)


if __name__ == "__main__":
    # REPLACE with your actual dataset path
    dataset_path = r"C:\Users\T2510597\Downloads\xaithesis\Diabetes_Final_Data_V2.csv"
    
    # Run the pipeline
    run_pipeline(data_path=dataset_path)
```

```python
"""
Counterfactual Analysis - Clinical Intervention Pathways (FIXED)
================================================================
Analyzes a randomly selected diabetic patient and generates personalized
counterfactual interventions with detailed clinical recommendations.

FIXED: Removed SMOTE, relaxed constraints to 50%, better error handling
"""

import pandas as pd
import numpy as np
import joblib
import dice_ml
from dice_ml import Dice
from sklearn.preprocessing import LabelEncoder
import random
import sys

# Load data and model
print("Loading data and model...")
df = pd.read_csv(r"C:\Users\T2510597\Downloads\xaithesis\Diabetes_Final_Data_V2.csv")

le = LabelEncoder()
df['gender_encoded'] = le.fit_transform(df['gender'])
df['target'] = le.fit_transform(df['diabetic'])

rf_model = joblib.load('diabetes_rf_model.pkl')

# Feature definitions
FEATURE_COLUMNS = [
    'age', 'gender_encoded', 'pulse_rate', 'systolic_bp', 'diastolic_bp',
    'glucose', 'height', 'weight', 'bmi', 'family_diabetes', 'hypertensive',
    'family_hypertension', 'cardiovascular_disease', 'stroke'
]

CATEGORICAL_FEATURES = [
    'gender_encoded','family_diabetes','hypertensive',
    'family_hypertension','cardiovascular_disease','stroke'
]

IMMUTABLE_FEATURES = ['age', 'gender_encoded', 'family_diabetes', 'family_hypertension', 'height']
MODIFIABLE_FEATURES = ['pulse_rate', 'systolic_bp', 'diastolic_bp', 'glucose', 
                       'weight', 'bmi', 'hypertensive', 'cardiovascular_disease', 'stroke']

# Helper functions for clinical assessment
def get_intervention_type(feature, original_val, target_val):
    """Determine specific intervention type based on feature and change."""
    interventions = {
        'glucose': {
            'primary': 'Glucose Management',
            'methods': ['Dietary modifications (low glycemic index diet)', 
                       'Oral hypoglycemic agents (Metformin)',
                       'Regular blood glucose monitoring']
        },
        'systolic_bp': {
            'primary': 'Blood Pressure Management',
            'methods': ['ACE inhibitors or ARBs',
                       'Lifestyle modifications (DASH diet)',
                       'Regular BP monitoring']
        },
        'diastolic_bp': {
            'primary': 'Blood Pressure Management',
            'methods': ['ACE inhibitors or ARBs',
                       'Sodium restriction',
                       'Stress management']
        },
        'weight': {
            'primary': 'Weight Management Program',
            'methods': ['Caloric restriction (500-750 kcal/day deficit)',
                       'Regular physical activity (150 min/week)',
                       'Behavioral counseling']
        },
        'bmi': {
            'primary': 'Weight Management Program',
            'methods': ['Structured diet plan',
                       'Exercise regimen',
                       'Weight tracking']
        },
        'pulse_rate': {
            'primary': 'Cardiovascular Fitness Program',
            'methods': ['Aerobic exercise (walking, cycling)',
                       'Gradual intensity progression',
                       'Beta-blockers if indicated']
        },
        'hypertensive': {
            'primary': 'Hypertension Treatment',
            'methods': ['Antihypertensive medication',
                       'Lifestyle modifications',
                       'Regular follow-up']
        }
    }
    
    return interventions.get(feature, {
        'primary': 'Lifestyle Modification',
        'methods': ['General health improvements']
    })

def assess_feasibility(feature, original_val, cf_val):
    """Assess feasibility with detailed timeframe and difficulty."""
    if feature in CATEGORICAL_FEATURES:
        return {
            'score': 80,
            'level': 'High',
            'timeframe': '3-6 months',
            'difficulty': 'Moderate'
        }
    
    change_pct = abs((cf_val - original_val) / original_val * 100) if original_val != 0 else 0
    
    if change_pct < 10:
        return {
            'score': 90,
            'level': 'High',
            'timeframe': '1-3 months',
            'difficulty': 'Low'
        }
    elif change_pct < 20:
        return {
            'score': 75,
            'level': 'High',
            'timeframe': '3-6 months',
            'difficulty': 'Moderate'
        }
    elif change_pct < 30:
        return {
            'score': 60,
            'level': 'Moderate',
            'timeframe': '6-12 months',
            'difficulty': 'Moderate'
        }
    else:
        return {
            'score': 40,
            'level': 'Low',
            'timeframe': '12+ months',
            'difficulty': 'High'
        }

def assess_clinical_impact(feature):
    """Assess clinical impact of changing this feature."""
    impact_map = {
        'glucose': 'Very High',
        'systolic_bp': 'High',
        'diastolic_bp': 'High',
        'hypertensive': 'High',
        'weight': 'Moderate',
        'bmi': 'Moderate',
        'cardiovascular_disease': 'High',
        'pulse_rate': 'Low',
        'stroke': 'High'
    }
    return impact_map.get(feature, 'Low')

# Prepare data for DiCE
print("\nPreparing data for counterfactual generation...")
df_dice = df.drop(columns=['gender','diabetic'], errors='ignore')
df_dice = df_dice[FEATURE_COLUMNS + ['target']]

for col in CATEGORICAL_FEATURES:
    df_dice[col] = df_dice[col].astype(str)

continuous_features = [f for f in FEATURE_COLUMNS if f not in CATEGORICAL_FEATURES]
d = dice_ml.Data(
    dataframe=df_dice,
    continuous_features=continuous_features,
    categorical_features=CATEGORICAL_FEATURES,
    outcome_name='target'
)

m = dice_ml.Model(model=rf_model, backend='sklearn')
dice_explainer = Dice(d, m, method='random')

# ============================================================================
# FIXED: Use original data, find high-risk diabetic patients
# ============================================================================

# Get predictions on full dataset
X = df[FEATURE_COLUMNS]
y = df['target']

# Find diabetic patients with high predicted risk
probabilities = rf_model.predict_proba(X)[:, 1]
diabetic_patients = np.where((y == 1) & (probabilities > 0.6))[0]

if len(diabetic_patients) == 0:
    print(" No high-risk diabetic patients found. Trying all diabetic patients...")
    diabetic_patients = np.where(y == 1)[0]

if len(diabetic_patients) == 0:
    print(" No diabetic patients found in dataset!")
    sys.exit(1)

# Select ONE random high-risk patient
random_patient_idx = random.choice(diabetic_patients)
patient_risk = probabilities[random_patient_idx]

print(f"\nRandomly selected patient index: {random_patient_idx}")
print(f"Patient diabetes risk: {patient_risk:.1%}")
print("=" * 100)

# Get patient data from ORIGINAL dataset (not resampled)
query_instance = X.iloc[[random_patient_idx]]
query_df = pd.DataFrame(query_instance.values, columns=FEATURE_COLUMNS)

for col in CATEGORICAL_FEATURES:
    if col in query_df.columns:
        query_df[col] = query_df[col].astype(float).astype(int).astype(str)

# ============================================================================
# FIXED: Relaxed constraints to 50% (was 30%)
# ============================================================================

print("\nBuilding intervention constraints (50% change allowed)...")
permitted_range = {}
for feat in FEATURE_COLUMNS:
    if feat in CATEGORICAL_FEATURES:
        current_value = str(query_df[feat].iloc[0])
    else:
        current_value = float(query_df[feat].iloc[0])
    
    if feat in IMMUTABLE_FEATURES:
        if feat in CATEGORICAL_FEATURES:
            permitted_range[feat] = [current_value]
        else:
            permitted_range[feat] = [current_value, current_value]
        print(f"   {feat}: LOCKED at {current_value}")
    elif feat in CATEGORICAL_FEATURES:
        unique_vals = sorted(df_dice[feat].unique().tolist())
        if feat in ['hypertensive', 'cardiovascular_disease', 'stroke']:
            if current_value == '1':
                permitted_range[feat] = ['0', '1']
                print(f"   {feat}: Can improve from 1 to 0")
            else:
                permitted_range[feat] = ['0']
                print(f"   {feat}: Already optimal (0)")
        else:
            permitted_range[feat] = unique_vals
    else:
        data_min = float(df_dice[feat].min())
        data_max = float(df_dice[feat].max())
        max_change = abs(current_value * 0.50)  # FIXED: 50% instead of 30%
        lower_bound = max(data_min, current_value - max_change)
        upper_bound = min(data_max, current_value + max_change)
        permitted_range[feat] = [float(lower_bound), float(upper_bound)]
        print(f"   {feat}: {lower_bound:.1f} to {upper_bound:.1f} (50% from {current_value:.1f})")

# Generate counterfactuals with better error handling
print("\nGenerating counterfactual interventions...")
print("(This may take 10-30 seconds...)")

try:
    dice_cfs = dice_explainer.generate_counterfactuals(
        query_df,
        total_CFs=3,  # Reduced from 5 for better success rate
        desired_class='opposite',
        features_to_vary=MODIFIABLE_FEATURES,
        permitted_range=permitted_range
    )
    
    cf_df = dice_cfs.cf_examples_list[0].final_cfs_df
    
    if cf_df is None or len(cf_df) == 0:
        print("\n DiCE could not find counterfactuals within constraints.")
        print("\nPossible reasons:")
        print("  1. Patient is at very high risk (hard to reverse)")
        print("  2. Constraints too strict (try relaxing to 70%)")
        print("  3. Too many immutable features")
        print("\nSuggestion: Try a different patient or relax constraints further")
        sys.exit(1)
    
    print(f"\n Generated {len(cf_df)} counterfactual pathways")
    
    # Print patient profile
    print("\n" + "=" * 100)
    print("PATIENT CLINICAL PROFILE")
    print("=" * 100)
    
    original = query_instance.iloc[0]
    age = int(original['age'])
    gender = 'Male' if int(original['gender_encoded']) == 1 else 'Female'
    
    print(f"\nDemographics:")
    print(f"   Age: {age} years")
    print(f"   Gender: {gender}")
    print(f"   Family Diabetes History: {'Yes' if int(original['family_diabetes']) == 1 else 'No'}")
    
    print(f"\nClinical Measurements:")
    print(f"   Glucose: {original['glucose']:.2f} mmol/L {' HIGH' if original['glucose'] > 7.0 else ' Normal'}")
    print(f"   Systolic BP: {int(original['systolic_bp'])} mmHg")
    print(f"   Diastolic BP: {int(original['diastolic_bp'])} mmHg")
    print(f"   Pulse Rate: {int(original['pulse_rate'])} bpm")
    print(f"   Weight: {original['weight']:.1f} kg")
    print(f"   BMI: {original['bmi']:.1f} {' OBESE' if original['bmi'] > 30 else ' OVERWEIGHT' if original['bmi'] > 25 else ' Normal'}")
    print(f"   Height: {original['height']:.2f} m")
    
    print(f"\nMedical Conditions:")
    print(f"   Hypertensive: {'Yes ' if int(original['hypertensive']) == 1 else 'No '}")
    print(f"   Cardiovascular Disease: {'Yes ' if int(original['cardiovascular_disease']) == 1 else 'No '}")
    print(f"   Stroke History: {'Yes ' if int(original['stroke']) == 1 else 'No '}")
    
    print(f"\n{'Current Status:':<20} Diabetic (Risk: {patient_risk:.1%})")
    print(f"{'Goal:':<20} Achieve Non-Diabetic Classification")
    
    # Analyze counterfactuals
    cf_analysis = []
    
    for cf_idx, cf_row in cf_df.iterrows():
        changes = []
        feasibility_scores = []
        impact_scores = []
        
        for feat in MODIFIABLE_FEATURES:
            if feat in cf_row.index and feat in original.index:
                if feat in CATEGORICAL_FEATURES:
                    orig_val = str(int(float(original[feat])))
                    cf_val = str(int(float(cf_row[feat])))
                else:
                    orig_val = float(original[feat])
                    cf_val = float(cf_row[feat])
                
                if orig_val != cf_val:
                    intervention_info = get_intervention_type(feat, orig_val, cf_val)
                    
                    if feat in CATEGORICAL_FEATURES:
                        change_desc = f"{feat.replace('_', ' ').title()}"
                        change_detail = f"{'No' if orig_val == '0' else 'Yes'}  {'No' if cf_val == '0' else 'Yes'}"
                        feasibility_data = assess_feasibility(feat, 1, 0)
                    else:
                        change_pct = ((cf_val - orig_val) / orig_val * 100) if orig_val != 0 else 0
                        change_desc = f"{feat.replace('_', ' ').title()}"
                        change_detail = f"{orig_val:.1f}  {cf_val:.1f} ({change_pct:+.0f}%)"
                        feasibility_data = assess_feasibility(feat, orig_val, cf_val)
                    
                    impact = assess_clinical_impact(feat)
                    
                    changes.append({
                        'feature': change_desc,
                        'change': change_detail,
                        'intervention_type': intervention_info['primary'],
                        'methods': intervention_info['methods'],
                        'feasibility': feasibility_data['level'],
                        'difficulty': feasibility_data['difficulty'],
                        'timeframe': feasibility_data['timeframe'],
                        'impact': impact
                    })
                    
                    feasibility_scores.append(feasibility_data['score'])
                    impact_scores.append({'Very High': 4, 'High': 3, 'Moderate': 2, 'Low': 1}[impact])
        
        if changes:
            avg_feasibility = np.mean(feasibility_scores)
            if avg_feasibility >= 70:
                overall_feas = 'High'
            elif avg_feasibility >= 50:
                overall_feas = 'Moderate'
            else:
                overall_feas = 'Low'
            
            avg_impact = np.mean(impact_scores)
            if avg_impact >= 3.5:
                overall_impact = 'Very High'
            elif avg_impact >= 2.5:
                overall_impact = 'High'
            elif avg_impact >= 1.5:
                overall_impact = 'Moderate'
            else:
                overall_impact = 'Low'
            
            if overall_feas == 'High' and overall_impact in ['Very High', 'High']:
                priority = 1
            elif overall_feas in ['High', 'Moderate'] and overall_impact in ['High', 'Moderate']:
                priority = 2
            else:
                priority = 3
            
            cf_analysis.append({
                'CF_ID': f'CF{cf_idx + 1}',
                'Changes': changes,
                'Num_Changes': len(changes),
                'Feasibility': overall_feas,
                'Clinical_Impact': overall_impact,
                'Priority': priority
            })
    
    # Sort by priority
    cf_analysis_sorted = sorted(cf_analysis, key=lambda x: (x['Priority'], -x['Num_Changes']))
    
    # SAVE DETAILED OUTPUT TO FILE
    output_file = 'counterfactual_analysis_report.txt'
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write("=" * 100 + "\n")
        f.write("COUNTERFACTUAL INTERVENTION PATHWAYS - DETAILED REPORT\n")
        f.write("=" * 100 + "\n\n")
        
        for cf in cf_analysis_sorted:
            f.write(f"\n{cf['CF_ID']} | Priority: {cf['Priority']} | Feasibility: {cf['Feasibility']} | Clinical Impact: {cf['Clinical_Impact']}\n")
            f.write("-" * 100 + "\n")
            
            for change in cf['Changes']:
                f.write(f"\n  Target: {change['feature']}\n")
                f.write(f"  Change Required: {change['change']}\n")
                f.write(f"  Intervention: {change['intervention_type']}\n")
                f.write(f"  Clinical Methods:\n")
                for method in change['methods']:
                    f.write(f"    - {method}\n")
                f.write(f"  Difficulty: {change['difficulty']} | Timeframe: {change['timeframe']} | Impact: {change['impact']}\n")
        
        # Clinical summary
        f.write("\n" + "=" * 100 + "\n")
        f.write("CLINICAL INTERPRETATION & RECOMMENDATIONS\n")
        f.write("=" * 100 + "\n\n")
        
        if cf_analysis_sorted:
            best_cf = cf_analysis_sorted[0]
            f.write(f"Recommended Clinical Pathway:\n")
            f.write(f"  Pathway: {best_cf['CF_ID']}\n")
            f.write(f"  Priority Level: {best_cf['Priority']}\n")
            f.write(f"  Overall Feasibility: {best_cf['Feasibility']}\n")
            f.write(f"  Expected Impact: {best_cf['Clinical_Impact']}\n")
            f.write(f"  Number of Interventions: {best_cf['Num_Changes']}\n")
    
    print(f"\n Detailed report saved to '{output_file}'")
    
    # Print SUMMARY to console
    print("\n" + "=" * 100)
    print("INTERVENTION SUMMARY (Top 3 Pathways)")
    print("=" * 100)
    
    for cf in cf_analysis_sorted[:3]:
        print(f"\n{cf['CF_ID']} | Priority {cf['Priority']} | {cf['Feasibility']} Feasibility | {cf['Clinical_Impact']} Impact | {cf['Num_Changes']} interventions")
        
        for i, change in enumerate(cf['Changes'][:3], 1):
            print(f"  {i}. {change['feature']}: {change['change']} - {change['intervention_type']} ({change['timeframe']})")
        
        if len(cf['Changes']) > 3:
            print(f"  ... and {len(cf['Changes']) - 3} more interventions")
    
    print(f"\n See '{output_file}' for complete detailed analysis")
    
    # Save to CSV
    output_rows = []
    for cf in cf_analysis_sorted:
        for change in cf['Changes']:
            output_rows.append({
                'Patient_ID': random_patient_idx,
                'Age': age,
                'Gender': gender,
                'Patient_Risk': f"{patient_risk:.1%}",
                'Counterfactual': cf['CF_ID'],
                'Priority': cf['Priority'],
                'Target_Feature': change['feature'],
                'Change_Required': change['change'],
                'Intervention_Type': change['intervention_type'],
                'Primary_Method': change['methods'][0] if change['methods'] else 'N/A',
                'Difficulty': change['difficulty'],
                'Timeframe': change['timeframe'],
                'Feasibility': cf['Feasibility'],
                'Clinical_Impact': cf['Clinical_Impact']
            })
    
    output_df = pd.DataFrame(output_rows)
    output_df.to_csv('counterfactual_analysis_detailed.csv', index=False)
    
    print("\n" + "=" * 100)
    print(" Analysis complete!")
    print("=" * 100)
    print("\nFiles created:")
    print("   counterfactual_analysis_report.txt - Full detailed report")
    print("   counterfactual_analysis_detailed.csv - Data for analysis")
    print("=" * 100)
        
except Exception as e:
    print(f"\n Error during counterfactual generation: {e}")
    print("\nDiagnostic Information:")
    print(f"   Patient index: {random_patient_idx}")
    print(f"   Patient risk: {patient_risk:.1%}")
    print(f"   Modifiable features: {len(MODIFIABLE_FEATURES)}")
    print(f"   Constraints applied: 50% change")
    print("\nTroubleshooting:")
    print("  1. Try running again (different patient may work better)")
    print("  2. Increase allowed change to 70% (edit max_change line)")
    print("  3. Check if patient has extreme values that can't be reversed")
    import traceback
    traceback.print_exc()
```

```python
"""
Visualize Class Weighting Impact on Loss Function
==================================================
Shows how algorithmic class weighting affects the model's learning
without modifying the actual data distribution.
"""

import matplotlib.pyplot as plt
import numpy as np

# Create figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# ============================================================================
# Subplot 1: Original Data Distribution (UNCHANGED)
# ============================================================================
ax1 = axes[0]

classes = ['Non-Diabetic', 'Diabetic']
counts = [5093, 344]
colors = ['#2ecc71', '#e74c3c']

bars1 = ax1.bar(classes, counts, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax1.set_ylabel('Number of Samples', fontsize=12, fontweight='bold')
ax1.set_title('(a) Original Data Distribution\n(UNCHANGED)', fontsize=13, fontweight='bold')
ax1.set_ylim(0, 6000)

# Add count labels
for bar, count in zip(bars1, counts):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 100,
             f'{count:,}',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

# Add ratio annotation
ratio = counts[0] / counts[1]
ax1.text(0.5, 0.95, f'Imbalance Ratio: {ratio:.1f}:1',
         transform=ax1.transAxes, ha='center', va='top',
         bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3),
         fontsize=10, fontweight='bold')

ax1.grid(axis='y', alpha=0.3, linestyle='--')

# ============================================================================
# Subplot 2: Effective Loss Weights (THE KEY VISUALIZATION)
# ============================================================================
ax2 = axes[1]

# Calculate class weights
weight_0 = 1.0  # Non-diabetic (majority)
weight_1 = counts[0] / counts[1]  # Diabetic (minority) = 15.78

weights = [weight_0, weight_1]
colors_weights = ['#3498db', '#e74c3c']

bars2 = ax2.bar(classes, weights, color=colors_weights, alpha=0.7, 
                edgecolor='black', linewidth=2)
ax2.set_ylabel('Loss Function Weight', fontsize=12, fontweight='bold')
ax2.set_title('(b) Algorithmic Class Weights\n(Applied to Loss Function)', 
              fontsize=13, fontweight='bold')
ax2.set_ylim(0, 20)

# Add weight labels
for bar, weight in zip(bars2, weights):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.3,
             f'{weight:.2f}',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

# Add explanation
ax2.text(0.5, 0.95, 'Diabetic errors penalized 15.78 more',
         transform=ax2.transAxes, ha='center', va='top',
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5),
         fontsize=10, fontweight='bold')

ax2.grid(axis='y', alpha=0.3, linestyle='--')

# ============================================================================
# Subplot 3: Effective Training Focus (Conceptual)
# ============================================================================
ax3 = axes[2]

# Show "effective importance" in training
effective_non_diabetic = 5093 * 1.0
effective_diabetic = 344 * weight_1

effective_values = [effective_non_diabetic, effective_diabetic]
colors_effective = ['#9b59b6', '#e67e22']

bars3 = ax3.bar(classes, effective_values, color=colors_effective, alpha=0.7,
                edgecolor='black', linewidth=2)
ax3.set_ylabel('Effective Training Emphasis', fontsize=12, fontweight='bold')
ax3.set_title('(c) Effective Model Focus\n(Weighted Learning)', 
              fontsize=13, fontweight='bold')
ax3.set_ylim(0, 6000)

# Add labels
for bar, eff_val in zip(bars3, effective_values):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 100,
             f'{eff_val:,.0f}',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

# Add balance annotation
ax3.text(0.5, 0.95, 'Balanced training emphasis',
         transform=ax3.transAxes, ha='center', va='top',
         bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5),
         fontsize=10, fontweight='bold')

ax3.grid(axis='y', alpha=0.3, linestyle='--')

# ============================================================================
# Overall title and layout
# ============================================================================
fig.suptitle('Algorithmic Class Weighting: Data Integrity Preserved, Model Behavior Adjusted',
             fontsize=16, fontweight='bold', y=1.02)

plt.tight_layout()

# Save figure
plt.savefig('class_weighting_visualization.png', dpi=300, bbox_inches='tight')
print(" Figure saved as 'class_weighting_visualization.png'")

plt.show()

# ============================================================================
# Generate LaTeX caption
# ============================================================================
caption = r"""
\caption{Algorithmic Class Weighting Strategy. 
(a) Original data distribution remains unchanged, preserving medical data integrity. 
(b) Loss function weights assigned to each class: minority class (Diabetic) receives 
15.78 higher penalty for misclassification. 
(c) Effective training emphasis after weighting, achieving balanced model focus without 
synthetic data generation. Unlike SMOTE which creates artificial samples, class weighting 
mathematically adjusts the learning algorithm while maintaining 100\% real patient data.}
\label{fig:class_weighting}
"""

print("\n" + "="*70)
print("LaTeX Caption for Thesis:")
print("="*70)
print(caption)
print("="*70)
```
